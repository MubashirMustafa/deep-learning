1. Linear Regression:
Real-life scenario: Predicting house prices based on features like area, number of bedrooms, and location.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Sample data
X = np.array([[1400], [1600], [1700], [1875], [1100]])
y = np.array([245000, 312000, 279000, 308000, 199000])

# Train the model
model = LinearRegression()
model.fit(X, y)

# Make predictions
predicted_prices = model.predict(X)

# Plotting
plt.scatter(X, y, color='blue')
plt.plot(X, predicted_prices, color='red')
plt.xlabel('Area (sq ft)')
plt.ylabel('Price ($)')
plt.title('House Prices vs. Area')
plt.show()
```

2. Logistic Regression:
Real-life scenario: Predicting whether an email is spam or not based on its content features.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression

# Sample data
X = np.array([[3], [4], [5], [6], [7]])
y = np.array([0, 0, 1, 1, 1])  # 0 represents non-spam, 1 represents spam

# Train the model
model = LogisticRegression()
model.fit(X, y)

# Make predictions
predicted_labels = model.predict(X)

# Plotting
plt.scatter(X, y, color='blue')
plt.plot(X, predicted_labels, color='red', marker='x')
plt.xlabel('Email Length')
plt.ylabel('Spam (1) / Non-Spam (0)')
plt.title('Spam Classification')
plt.show()
```

3. Decision Trees:
Real-life scenario: Predicting whether a loan applicant is likely to default based on features like income, credit score, and loan amount.

```python
from sklearn.datasets import load_iris
from sklearn import tree

# Sample data
iris = load_iris()
X = iris.data
y = iris.target

# Train the model
model = tree.DecisionTreeClassifier()
model.fit(X, y)

# Plotting
fig, ax = plt.subplots(figsize=(12, 12))
tree.plot_tree(model, filled=True, feature_names=iris.feature_names, class_names=iris.target_names, ax=ax)
plt.show()
```

4. Random Forest:
Real-life scenario: Predicting customer churn in a telecom company based on various customer attributes.

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import make_classification

# Sample data
X, y = make_classification(n_samples=1000, n_features=10, n_classes=2, random_state=42)

# Train the model
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X, y)

# Feature importance
importances = model.feature_importances_

# Plotting
plt.bar(range(X.shape[1]), importances)
plt.xticks(range(X.shape[1]), ['Feature ' + str(i+1) for i in range(X.shape[1])])
plt.xlabel('Feature')
plt.ylabel('Importance')
plt.title('Feature Importance in Random Forest')
plt.show()
```

5. Support Vector Machines (SVM):
Real-life scenario: Classifying images as either cats or dogs based on features extracted from the images.

```python
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
import numpy as np
import matplotlib.pyplot as plt

# Sample data with two features
X, y = make_classification(n_samples=100, n_features=2, n_classes=2, random_state=42, n_informative=2, n_redundant=0, n_clusters_per_class=1)

# Train the model
model = SVC(kernel='linear')
model.fit(X, y)

# Plotting
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='viridis')

# Plot decision boundary
ax = plt.gca()
xlim = ax.get_xlim()
ylim = ax.get_ylim()
xx, yy = np.meshgrid(np.linspace(xlim[0], xlim[1], 50), np.linspace(ylim[0], ylim[1], 50))
Z = model.decision_function(np.c_[xx.ravel(), yy.ravel()]).reshape(xx.shape)

plt.contour(xx, yy, Z, colors='k', levels=[-1, 0, 1], alpha=0.5, linestyles=['--', '-', '--'])
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('SVM Decision Boundary')
plt.show()
```
